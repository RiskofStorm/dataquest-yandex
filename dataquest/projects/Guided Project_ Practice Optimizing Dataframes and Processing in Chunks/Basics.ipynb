```python
import pandas as pd
import pprint as pp 
pd.options.display.max_columns = 99
pp = pp.PrettyPrinter(depth=10)
```

```python
#raw_data = pd.read_csv("loans_2007.csv")
data = pd.read_csv("loans_2007.csv", chunksize=3100)
#data["id"].astype("int", errors="ignore")
#data.dropna(thresh=len(data.columns),inplace=True)

for i in data:
    print("memory :",(i.memory_usage(deep=True).sum())/1024**2)
    print("number of numeric types: ", len(i.select_dtypes(include=["int", "float"]).dtypes))
    print("number of string types: ", len(i.select_dtypes(include=["object"]).dtypes))
    print("total columns :", len(i.columns),"\n")
    break
```

```python
unique_str = dict()
for i in data:
    for j in i.select_dtypes(include=["object"]).columns:
        col_unique = i[j].value_counts()
        if j in unique_str:
            unique_str[j].append(col_unique)
        else:
            unique_str[j] = [col_unique]

p_unique = dict()
for col in unique_str:
    pre_proc = pd.concat(unique_str[col])
    p_unique[col] = pre_proc.groupby(pre_proc.index).sum()
p_unique
```

```python
data = pd.read_csv("loans_2007.csv", chunksize=3100)
for i in data:
    str_dset = i.select_dtypes(include=["object"])
    break
    
    

to_cat = ["zip_code","emp_length","term","grade","sub_grade","home_ownership","verification_status",
          "loan_status","addr_state","initial_list_status","application_type"]


cls = pd.read_csv("loans_2007.csv", chunksize=3100)
temp_cls= list()
for i in cls:
    
    for j in ["int_rate","revol_util"]:
        i[j] = i[j].astype(str).str.replace("%","")
    
    for k in to_cat:
        i[k] = i[k].astype("category").cat.codes
        
    for g in ["last_pymnt_d","last_credit_pull_d","issue_d","earliest_cr_line"]:
        i[g] = pd.to_datetime(i[g])
    temp_cls.append(i)

proc_cls = pd.concat(temp_cls)
proc_cls.head(10)
```

```python
pp.pprint(proc_cls.shape[0])
proc_cls.drop(['emp_title','title'],axis=1, inplace=True)

for col in ["issue_d","last_credit_pull_d"]:
    proc_cls[col].fillna(method="ffill",inplace=True)
    proc_cls[col].isna().sum()
```
