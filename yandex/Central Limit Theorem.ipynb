```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Pareto distribution
https://ru.wikipedia.org/wiki/Распределение_Парето

```python
b = 5
data = np.linspace(sts.pareto.ppf(0.01, b), sts.pareto.ppf(0.99, b), 1000)
f_data = sts.pareto.pdf(data,b)

plt.figure(figsize=(8,5),dpi=80)
plt.plot(data, f_data)
plt.grid(True)
plt.title("Pareto probability density fucntion", size=15)
plt.xlabel("Pr(X==x)")
plt.ylabel("x")
plt.show()

plt.figure(figsize=(8,5),dpi=80)
plt.title("Pareto PDF histogram", size=15)
plt.hist(f_data)
plt.xlabel("")
plt.ylabel("almount")
plt.show()
```

After we're done with Pareto distribution let's start get sample of data from it.  

После постройки Распределения Парето, давайте начнем делать выборки из него.

```python
# calc

def pareto_approx(n,f_data):
    sample_n = list()
    
    for i in range(1000):
        sample = f_data[np.random.randint(1,1000,n)]
        sample = sum(sample)/n
        sample_n.append(sample)
        
    plt.figure(figsize=(8,5), dpi=80)
    plt.title(f"n = {n} approximation", size=15)
    plt.hist(sample_n, density=True)
    sns.kdeplot(sample_n)
    plt.xlabel("sample mean")
    plt.ylabel("almount")
    plt.show()
    
    
n = [5,10,50]
for i in n:
    pareto_approx(i,f_data)
```

As we can see increasing number of n-sample in Pareto distribution we approximate our total sample to normal.
This is magic of Law of large numbers.

Как мы видим при увеличении n выборки из Распределения Парето, мы приближаемся к нормальному распределению. Но из-за того, что распределение Парето скошено влево, нам необходимо число n больше чем в других стандартных случаях.

```python
pareto_approx(5000,f_data)
```

When we have in a sample approximation to normal distribution, then we could say that we approximate to representative
sample, which is useful as a low-weight version of general population for further data analysis and machine learning.

Когда у нас в выборке есть приближение к нормальному распределению, значит мы приблизись к репрезентативной выборке, тем самым уменьшили кол-во данных для дальнейшей обработки, машинного обучения.
