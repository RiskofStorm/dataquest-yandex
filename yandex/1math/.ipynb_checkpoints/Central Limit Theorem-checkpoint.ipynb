### *Pareto mean*
\begin{cases}\infty &{\text{for }}\alpha \leq 1\\{\frac {\alpha \,x_{\mathrm {m} }}{\alpha -1}}&{\text{for }}\alpha >1\end{cases}
### *Pareto Variance*
\begin{cases}\infty &{\text{for }}\alpha \leq 2\\{\frac {x_{\mathrm {m} }^{2}\alpha }{(\alpha -1)^{2}(\alpha -2)}}&{\text{for }}\alpha >2\end{cases}

```python
def pareto_approx(n,b):
    
    # Counting random sample
    pareto_rv = sts.pareto(b)
    n_data = [np.average(pareto_rv.rvs(n)) for _ in range(1000)]
    
    # Counting theoretical data
    x = np.linspace(0,10,1000)

    mean = b / (b-1)
    sd = np.sqrt(((1 / ((b-1) ** 2)) * (b / (b-2))) / n)
    
    t_data = sts.norm(mean,sd).pdf(x)
    
    plt.figure(figsize=(8,5), dpi=100)
    plt.hist(n_data, bins=100,range=(1,10), density=True, label='Sample mean density')
    plt.plot(x, t_data, label= "Theoretical mean density")
    plt.legend(loc="best")
    plt.title(f" n = {n}", size=15)
    plt.xlabel("sample mean")
    plt.ylabel("almount")
    plt.show()
    
    
n = [5,10,50]
for i in n:
    pareto_approx(i,3)
```


Как мы видим при увеличении n выборки из Распределения Парето, мы приближаемся к нормальному распределению. Но из-за того, что распределение Парето скошено влево, нам необходимо число n больше чем в других стандартных случаях.
