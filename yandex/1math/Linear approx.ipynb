```python
import math
import numpy as np
import matplotlib.pyplot  as plt
from __future__ import division
```

### f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
x1 = (1, 15)
x2 = (1, 8, 15)
x3 = (1, 4, 10, 15)


def approx(x):
    
    def f(x):
        return  math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

    
    a_matrix = [[i**n for n in range(len(x))]for i in x]
    b_vector = [round(f(i),2) for i in x]

    a_matrix = np.array(a_matrix)
    b_vector = np.array(b_vector)

    result = np.linalg.solve(a_matrix,b_vector)
    print(a_matrix)
    print(b_vector)
    print(result)

    plt.plot(sorted(result), b_vector)
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
#     plt.axis([0,20,0,20])
    plt.grid(True)
    plt.show()
    
    return result
    

approx(x1)
approx(x2)
result = approx(x3)
```

```python
ans = [str(round(i,2)) for i in result]
print(tuple(ans))
with open('result.txt', 'w') as wf:
    for i in ans:
        wf.write(i + ' ')
```
